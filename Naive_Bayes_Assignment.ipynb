Now that we have joined the 2 sets and performed train test split, now performing Naive Bayes Classifier.

##Naive Bayes Classifier

```python
!pip install sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report as report
```

**Gaussian Naive Bayes**

```python
GNB_model=GaussianNB()
GNB_model.fit(x_train,y_train)
GNB_y_pred=GNB_model.predict(x_test)
#Accuracy
print("Model Accuracy for GaussianNB: {0:0.4f}".format(accuracy_score(y_test,GNB_y_pred)))
```

**Bernoulli Naive Bayes**

```python
BNB_model=BernoulliNB()
BNB_model.fit(x_train,y_train)
BNB_y_pred=BNB_model.predict(x_test)
print("Model Accuracy for BernoulliNB: {0:0.4f}".format(accuracy_score(y_test,BNB_y_pred)))
```

**Categorical Naive Bayes**

```python
CNB_model=CategoricalNB()
CNB_model.fit(x_train,y_train)
CNB_y_pred=CNB_model.predict(x_test)
print("Model Accuracy for CategoricalNB: {0:0.4f}".format(accuracy_score(y_test,CNB_y_pred)))
```

**Classification Report**

```python
print("GaussianNB Model","\n",report(y_test,GNB_y_pred))
print("BernoilliNB Model","\n",report(y_test,BNB_y_pred))
print("CategoricalNB Model","\n",report(y_test,CNB_y_pred))
```

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix as cm

CNB_y_pred=CNB_model.predict(x_test)
print(cm(y_test,CNB_y_pred))
```

```python
cm_df=cm(y_test,CNB_y_pred)
class_label=["NO","YES"]

df_cm=pd.DataFrame(cm_df,index=class_label,columns=class_label)
sns.heatmap(df_cm,annot=True,fmt="d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.show()
```

**Finding AUC value & plotting ROC curve**

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr,tpr,thresholds=roc_curve(y_test,CNB_model.predict_proba(x_test)[:,1])
auc=roc_auc_score(y_test,CNB_y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area=%0.2f)'%auc)
plt.plot([0, 1],[0, 1],'k--')

plt.title('ROC Curve')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('AUC-ROC Score:',auc) 
```

```python
print('Training score: {:.2f}%'.format(CNB_model.score(x_train, y_train)*100))

print('Test score: {:.2f}%'.format(CNB_model.score(x_test, y_test)*100))
```

As the train & test scores doesn't have much difference & can be compared, this is a sign of proper fitting.
